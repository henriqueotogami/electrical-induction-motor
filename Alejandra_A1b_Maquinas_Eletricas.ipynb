- Vth = Tens√£o de Thev√©nin no rotor | (V) - Volts

- X1 = Indut√¢ncia do rotor | (‚Ñ¶) - Ohms

- Xth = Indut√¢ncia de Thev√©nin | (‚Ñ¶) - Ohms
